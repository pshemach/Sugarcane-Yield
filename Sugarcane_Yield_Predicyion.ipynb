```python
generate_df = pd.read_csv('https://raw.githubusercontent.com/pshemach/Sugarcane-Yield/main/Yield_generated_data.csv')
generate_df
```

```python
generate_df.describe()
```

```python
generate_df.columns
```

```python
X_new = generate_df[['Maximum_Temp', 'Minimum_temp', 'Relative Humidity_AVG', 'Rain fall_mm',
       'Wind velocity_km/hr', 'Sunshine_hr']]
y_new = generate_df['Yield (t/ha)']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size=0.4, random_state=100)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

R2 =[]
rmse =[]
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=100)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

print(rf_pred)

rf_r2 = r2_score(y_test,rf_pred)
R2.append(rf_r2)

rf_mse = mean_squared_error(y_test,rf_pred)
rf_rmse =np.sqrt(rf_mse)
rmse.append(rf_rmse)
```

```python
rf_r2,rf_rmse
```

```python
from sklearn.svm import SVR

svr_model = SVR(kernel='rbf', C=1, epsilon=0.1)

svr_model.fit(X_train, y_train)

svr_pred = svr_model.predict(X_test)

print(svr_pred)

svr_r2 = r2_score(y_test,svr_pred)
R2.append(svr_r2)

svr_mse = mean_squared_error(y_test,svr_pred)
svr_rmse =np.sqrt(svr_mse)
rmse.append(svr_rmse)
```

```python
svr_r2,svr_rmse
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=100)

gbr_model.fit(X_train, y_train)

gbr_pred = gbr_model.predict(X_test)

print(gbr_pred)

gbr_r2 = r2_score(y_test,gbr_pred)
R2.append(gbr_r2)

gbr_mse = mean_squared_error(y_test,gbr_pred)
gbr_rmse =np.sqrt(gbr_mse)
rmse.append(gbr_rmse)
```

```python
gbr_r2, gbr_rmse
```

```python
R2, rmse
```

```python
metrics = pd.DataFrame([R2,rmse], columns = ['RandomForestRegressor','SVR','GradientBoostingRegressor'], index=['R2','rmse']).transpose()
metrics
```

```python
sns.scatterplot(data=metrics, x = metrics.index , y='R2')
sns.lineplot(data=metrics, x = metrics.index , y='R2', color='r')
plt.show()
```

```python
sns.scatterplot(data=metrics, x = metrics.index , y='rmse')
sns.lineplot(data=metrics, x = metrics.index , y='rmse', color='g')
plt.show()
```

```python
sns.barplot(data=metrics, x = metrics.index , y='R2')
plt.show()
```
